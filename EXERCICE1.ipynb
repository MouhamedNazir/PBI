6.	Ajouter une colonne "Email" : créer une adresse sous la forme prenom.nom@usmba.ac.ma à partir des colonnes Nom et Prénom en supprimant les espaces, apostrophes (') et tirets (-), puis en concaténant Nom et Prénom avec un point, en majuscules, et en ajoutant @usmba.ac.ma (ex : EL MOQADEM + AYMANE MOUBARIK → elmoqadem.aymanemoubarik@usmba.ac.ma, N'MAR + HAJAR → nmar.hajar@usmba.ac.ma, ET-TAALABI + SEFIANE → ettaalabi.sefiane@usmba.ac.ma).


```python
def nettoyer(text):
    text = str(text).lower()
    text = text.replace(" ","").replace("'","").replace("-","")

    return text

df["Email"] = df["Nom"].apply(nettoyer)+"."+df["Prenom"].apply(nettoyer)+"@usmba.ac.ma"
                                                               
    
```
